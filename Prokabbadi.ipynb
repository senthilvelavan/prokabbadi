```python
from sklearn.model_selection import train_test_split
x_train,x_valtest,y_train,y_valtest = train_test_split(x,y,test_size=0.3,random_state=29)
x_val,x_test,y_val,y_test = train_test_split(x_valtest,y_valtest,test_size=0.5,random_state=29)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc_pl = GradientBoostingClassifier()
gbc_pl.fit(x_train,y_train)
y_val_pred = gbc_pl.predict(x_val)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_val,y_val_pred))
print(confusion_matrix(y_val,y_val_pred))
```

```python
y_test_pred = gbc_pl.predict(x_test)
print(classification_report(y_test,y_test_pred))
print(confusion_matrix(y_test,y_test_pred))
```

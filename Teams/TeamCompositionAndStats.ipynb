```python
def getTeamStats(teamPlayersLink,teamName,playersName,playersRole,team):
    driver = webdriver.Chrome('D:\BITS\Prokabbadi\chromedriver_win32\chromedriver.exe')
    try:

        #driver.get("https://www.prokabaddi.com/teams/bengal-warriors-profile-4/players")
        driver.get(teamPlayersLink)
        import time
        time.sleep(2)
        soup = BeautifulSoup(driver.page_source,"lxml")
        #print(soup.prettify())
        player_names = soup.find(class_="si-page-plyrWrp")
        #player_names = soup.find(class_='si-innerWrp')
        #print(player_names)
        players = player_names.find_all(class_="si-plyrBox")
        #print(players)
        
        for player in players:
            #print('---------------------player#',count,'----------------------')
            
            #playerName = player.find(class_='si-plyrName')
            name = player.find(class_="si-txt1").get_text().strip()+' '+player.find(class_="si-txt2").get_text().strip()
            print(name)
            print(player.find(class_='si-plyrRole').get_text().strip())
            playersName.append(name)
            playersRole.append(player.find(class_='si-plyrRole').get_text().strip())
            team.append(teamName)
            #count = count + 1
            #print(player)
        print(len(playersName))
        print(len(playersRole))
        return playersName,playersRole,team
        
    except:
        print(traceback.format_exc())
        driver.close()
```

```python
def getTeamStats(teamStatsLink,teamName,overall_p,attack_p,defence_p,overall_s,attack_s,defence_s):
    driver = webdriver.Chrome('D:\BITS\Prokabbadi\chromedriver_win32\chromedriver.exe')
    try:

        #driver.get("https://www.prokabaddi.com/teams/bengal-warriors-profile-4/players")
        driver.get(teamStatsLink)
        import time
        time.sleep(2)
        soup = BeautifulSoup(driver.page_source,"lxml")
        overall = soup.find(class_="si-tbl-wraper si-overall")
        overall_tbl_data = overall.find_all(class_ = 'si-data-block')
        print(len(overall_tbl_data))
        for data in overall_tbl_data:
            print(data.get_text().strip())
            overall_p.append(data.get_text().strip())
            
        attack = soup.find(class_="si-tbl-wraper si-attack")
        attack_tbl_data = attack.find_all(class_ = 'si-data-block')
        print(len(attack_tbl_data))
        for data in attack_tbl_data:
            print(data.get_text().strip())
            attack_p.append(data.get_text().strip())
            
        defence = soup.find(class_="si-tbl-wraper si-defence")
        defence_tbl_data = defence.find_all(class_ = 'si-data-block')
        print(len(defence_tbl_data))
        for data in defence_tbl_data:
            print(data.get_text().strip())
            defence_p.append(data.get_text().strip())
            
        season_header = soup.find(class_='si-section-data si-right').find(class_='si-tbl-wraper si-tbl-header')
        seasons = season_header.find_all(class_='si-tbl-data')
        print(len(seasons))
        for season in seasons:
            print(season.find(class_='si-data-block').get_text().strip())
            
        all_stats = soup.find(class_='si-section-data si-right').find_all(class_='si-tbl-wraper')
        oad_count = 0
        for stat in all_stats:
            
                print('oad_count is ',oad_count)
                if(oad_count == 0):
                    print('length of overall is ',len(overall_s))
                    list_mtrcs = overall_s
                elif(oad_count == 1):
                    print('length of attack is ',len(attack_s))
                    list_mtrcs = attack_s
                elif(oad_count == 2):
                    print('length of defence is ',len(defence_s))
                    list_mtrcs = defence_s
                tbl_datas = stat.find_all(class_='si-tbl-data')
                breakLoop = True
                if(tbl_datas[0].find(class_='si-data-block').get_text().strip() == 'Overall'):
                    breakLoop = False
                    continue;
                oad_metrics_count = 0
                for tbl_data in tbl_datas:
                    data_blocks = tbl_data.find_all(class_='si-data-block')
                    #print(len(data_blocks))

                    for data_block in data_blocks:
                        print('oad_metrics_count is ',oad_metrics_count)
                        print(data_block.get_text().strip())
                        list_mtrcs[oad_metrics_count].append(data_block.get_text().strip())


                    oad_metrics_count = oad_metrics_count + 1
                if(oad_count==0):
                    print('assign overall_ss to list_mtrcs')
                    overall_ss = list_mtrcs
                elif(oad_count == 1):
                    print('assign attack_ss to list_mtrcs')
                    attack_ss = list_mtrcs
                elif(oad_count == 2):
                    print('assign defence_ss to list_mtrcs')
                    defence_ss = list_mtrcs
                if(breakLoop == True):
                    oad_count = oad_count + 1
        return overall_p,attack_p,defence_p,overall_ss,attack_ss,defence_ss
            
        
    except:
        print(traceback.format_exc())
        driver.close()
```
